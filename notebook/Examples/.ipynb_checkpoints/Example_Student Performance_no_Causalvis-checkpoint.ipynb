```python
import json
import pandas as pd
```

### Load Data

```python
data = pd.read_csv('./data/student-mat.csv', delimiter=';')
data.head(5)
```

```python
drop_col = ['school','sex','age','Mjob', 'Fjob','reason','guardian']
data = data.drop(columns=drop_col)
data.head(5)
```

# Causal Structure Modeling

```python
import numpy as np

struct_data = data.copy()
non_numeric_columns = list(struct_data.select_dtypes(exclude=[np.number]).columns)

print(non_numeric_columns)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in non_numeric_columns:
    struct_data[col] = le.fit_transform(struct_data[col])

struct_data.head(5)
```

### Creating and Editing the DAG

We use Causalnex to perform causal discovery on our data set and create an initial "best guess" of the DAG.
This is then manually refined.

```python
import warnings
from causalnex.structure import StructureModel

warnings.filterwarnings("ignore")  # silence warnings

sm = StructureModel()
```

```python
from causalnex.structure.notears import from_pandas

sm = from_pandas(struct_data)
sm.remove_edges_below_threshold(0.8)
```

### Visualizing the initial graph discovered by Causalnex

```python
from IPython.display import Image
from causalnex.plots import plot_structure, NODE_STYLE, EDGE_STYLE

viz = plot_structure(
    sm,
    graph_attributes={"scale": "0.5"},
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
    prog='fdp',
)
Image(viz.draw(format='png'))
```

### Adding missing edges

```python
sm.add_edge("famsup", "absences")
sm.add_edge("health", "absences")
sm.add_edge("health", "G1")
sm.add_edge("Medu", "G1")
sm.add_edge("Medu", "higher")
sm.add_edge("internet", "G1")
sm.add_edge("address", "absences")
sm.add_edge("Pstatus", "famrel")
sm.add_edge("paid", "G1")
sm.add_edge("absences", "G1")

viz = plot_structure(
    sm,
    graph_attributes={"scale": "0.5"},
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
    prog='fdp',
)
Image(viz.draw(format='png'))
```

### Removing edges

```python
sm.remove_edge("higher", "traveltime")
sm.remove_edge("Dalc", "Walc")
sm.remove_edge("Walc", "absences")
sm.remove_edge("higher", "famrel")
sm.remove_edge("romantic", "absences")
sm.remove_edge("address", "G1")
sm.remove_edge("schoolsup", "absences")
sm.remove_edge("higher", "studytime")
sm.remove_edge("higher", "Medu")

viz = plot_structure(
    sm,
    graph_attributes={"scale": "0.5"},
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
    prog='fdp',
)
Image(viz.draw(format='png'))
```

### Remove unconnected nodes

```python
sm = sm.get_largest_subgraph()
viz = plot_structure(
    sm,
    graph_attributes={"scale": "0.5"},
    all_node_attributes=NODE_STYLE.WEAK,
    all_edge_attributes=EDGE_STYLE.WEAK,
)
Image(viz.draw(format='png'))
```

### Identifying control variables

Once we are satisfied with the causal structure, we can identify the treatment and outcome variables. In this case, we are interested in `G1` as the outcome variable, while `absences` is the treatment variable of interest.

From the visualization, we see that the confounds are: 'Pstatus', 'failures', 'internet', 'health', 'famsup', 'Medu'

The prognostic factors are: 'studytime', 'schoolsup', 'paid', 'higher'

```python
confounds = ['Pstatus', 'failures', 'internet', 'health', 'famsup', 'Medu']
```

```python
prognostics = ['studytime', 'schoolsup', 'paid', 'higher']
```

# Cohort Construction and Refinement

```python
data.shape
```

```python
a = data["absences"]
a = a >= a.median()
a = a.astype(int)
```

```python
y = data["G1"]
```

```python
covariates = confounds + prognostics

X = data[covariates]
X.dtypes
```

```python
X = data.loc[:, covariates]
X = pd.get_dummies(X, drop_first=True)
X.shape
```

```python
from sklearn.linear_model import LogisticRegression
from causallib.estimation import IPW

lr = LogisticRegression(solver='lbfgs', max_iter=1000)
ipw = IPW(lr)
```

```python
ipw.fit(X, a)
```

```python
propMatrix = ipw.compute_propensity_matrix(X, a).to_dict(orient="records")
```

### Matching

```python
from causallib.estimation import IPW, Matching
from causallib.preprocessing.transformers import PropensityTransformer
from causallib.preprocessing.transformers import MatchingTransformer
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
matcher = MatchingTransformer(with_replacement=True, n_neighbors=1, caliper=2)
matcher.fit(X, a, y)
Xm, am, ym = matcher.transform(X, a, y)

Xm.reset_index(drop=True, inplace=True)
am.reset_index(drop=True, inplace=True)
ym.reset_index(drop=True, inplace=True)

def learner(): return LogisticRegression(solver="liblinear",
                                         max_iter=5000,
                                         class_weight="balanced")

ipw = IPW(learner=learner())
ipw.fit(Xm, am)
propMatrixm = ipw.compute_propensity_matrix(Xm, am).to_dict(orient="records")
```

```python
Xm.shape
```

### Visualize Propensity Score Plots for First Matched Cohort

```python
propTreatment = [propMatrixm[ind][0] for ind, assignment in am.iteritems() if assignment == 0]
propControl = [propMatrixm[ind][1] for ind, assignment in am.iteritems() if assignment == 1]
```

```python
import matplotlib.pyplot as plt

f, (ax1, ax2) = plt.subplots(1, 2)

nt, binst, patchest = ax1.hist(propTreatment, 20, range=[0, 1], density=True, facecolor='orchid', alpha=0.75)
ax1.title.set_text('Treatment')

nc, binsc, patchesc = ax2.hist(propControl, 20, range=[0, 1], density=True, facecolor='steelblue', alpha=0.75)
ax2.title.set_text('Control')

plt.show()
```

```python
matcher = MatchingTransformer(with_replacement=True, n_neighbors=1, caliper=1)
matcher.fit(X, a, y)
Xm2, am2, ym2 = matcher.transform(X, a, y)

Xm2.reset_index(drop=True, inplace=True)
am2.reset_index(drop=True, inplace=True)
ym2.reset_index(drop=True, inplace=True)

def learner(): return LogisticRegression(solver="liblinear",
                                         max_iter=5000,
                                         class_weight="balanced")

ipw = IPW(learner=learner())
ipw.fit(Xm2, am2)
propMatrixm2 = ipw.compute_propensity_matrix(Xm2, am2).to_dict(orient="records")
```

```python
Xm2.shape
```

### Visualize Propensity Score Plots for Second Matched Cohort

```python
propTreatment2 = [propMatrixm2[ind][0] for ind, assignment in am2.iteritems() if assignment == 0]
propControl2 = [propMatrixm2[ind][1] for ind, assignment in am2.iteritems() if assignment == 1]
```

```python
import matplotlib.pyplot as plt

f, (ax1, ax2) = plt.subplots(1, 2)

nt, binst, patchest = ax1.hist(propTreatment2, 20, range=[0, 1], density=True, facecolor='orchid', alpha=0.75)
ax1.title.set_text('Treatment')

nc, binsc, patchesc = ax2.hist(propControl2, 20, range=[0, 1], density=True, facecolor='steelblue', alpha=0.75)
ax2.title.set_text('Control')

plt.show()
```

```python
propensity_transform = PropensityTransformer(
    include_covariates=False, learner=learner())
matcher = Matching(propensity_transform=propensity_transform,
                   with_replacement=True, n_neighbors=1, knn_backend="sklearn")
matcher.fit(Xm2, am2, ym2)
matches = matcher._execute_matching(Xm2, am2)
```

```python
matcher.estimate_population_outcome(Xm2, am2)
```

```python
population_outcome = matcher.estimate_population_outcome(Xm2, am2)
ATE = population_outcome[1] - population_outcome[0]
ATE
```

# Treatment Effect Exploration

```python
ind_outcomes = matcher.estimate_individual_outcome(Xm2, am2).to_dict(orient="records")

ind_effects = []

for i in ind_outcomes:
    i_treatment = i[1]
    i_control = i[0]
    
    effect = i_treatment - i_control
    
    ind_effects.append(effect)
```

```python
indv_data = []
Xm_dict = Xm2.to_dict(orient="records")

for i in range(len(Xm_dict)):
    newDataInstance = Xm_dict[i]
    newDataInstance['treatment'] = int(am2[i])
    newDataInstance['outcome'] = int(ym2[i])
    newDataInstance['effect'] = int(ind_effects[i])
    indv_data.append(newDataInstance)
```

First visualize the distribution for the entire cohort

```python
ITE = [x["effect"] for x in indv_data]
```

```python
plt.violinplot([ITE], positions=[1], points=50, widths=0.3,
                     showmeans=True, showextrema=True, showmedians=True)
plt.title('Individual Treatment Effects for Cohort')
plt.show()
```

Subgroup by internet access to create 2 distributions.

```python
ITE_internet_0 = [x["effect"] for x in indv_data if x["internet_yes"] == 0]
ITE_internet_1 = [x["effect"] for x in indv_data if x["internet_yes"] == 1]
```

```python
plt.violinplot([ITE_internet_0, ITE_internet_1], positions=[0, 1], points=50, widths=0.3,
                     showmeans=True, showextrema=True, showmedians=True)
plt.title('Individual Treatment Effects')
plt.xticks([0, 1], ["internet_yes = 0", "internet_yes = 1"])
plt.show()
```

Subgroup by internet access and student health to get a faceted visualization.

```python
ITE_internet_0_health_low = [x["effect"] for x in indv_data if x["internet_yes"] == 0 and x["health"] < 3.52]
ITE_internet_0_health_high = [x["effect"] for x in indv_data if x["internet_yes"] == 0 and x["health"] > 3.52]

ITE_internet_1_health_low = [x["effect"] for x in indv_data if x["internet_yes"] == 1 and x["health"] < 3.52]
ITE_internet_1_health_high = [x["effect"] for x in indv_data if x["internet_yes"] == 1 and x["health"] > 3.52]

f, (ax1, ax2) = plt.subplots(1, 2)

ax1.violinplot([ITE_internet_0_health_low, ITE_internet_1_health_low], positions=[0, 1], points=50, widths=0.3,
                     showmeans=True, showextrema=True, showmedians=True)
ax1.title.set_text("health < 3.52")

ax2.violinplot([ITE_internet_0_health_high, ITE_internet_1_health_high], positions=[0, 1], points=50, widths=0.3,
                     showmeans=True, showextrema=True, showmedians=True)
ax2.title.set_text("health > 3.52")

plt.xticks([0, 1])

plt.show()
```
